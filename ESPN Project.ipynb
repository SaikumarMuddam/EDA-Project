```python
URL='https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_innings.html?class=3;id=30;type=team'
URL
```

```python
# Loading the WebPage in Memory using requests library

page = requests.get(URL)
page
```

```python
# Extracting the HTML Code of the WebPage

htmlcode = page.text
```

```python
htmlcode
```

```python
# Format the HTML code using bs4 library

soup = BeautifulSoup(htmlcode)
soup
```

```python
soup.find_all('table')
```

```python
print(soup.prettify())
```

```python

pl= soup.find('td', attrs={'class' : 'left'})

print(pl.text)
```

```python

pla= soup.find('table', attrs={'class' : 'engineTable'})

print(pla.text)
```

```python
a=[]
for i in soup.find_all('tr'):
    a.append(i.text)
a
```

```python
df = pd.DataFrame({'available' : a})
df 
```

```python
df=df.available.str.split('\n', expand=True)
df
```

```python
df1=df.dropna()
df1
```

```python
df3=df1.drop([0], axis=0)
df3
```

```python
#Cleaning the data

df3.columns=['0','Player','Runs','Balls','4s','6s','SR','7','Opposition','Ground','Match_Date','Scorecard','12']
```

```python
df3
```

```python
#Cleaning the data

df4=df3.drop(['0','7','12'], axis=1) 
df4
```

```python
#Cleaning the data

df4.Opposition = df4.Opposition.apply(lambda x : x.replace('v', ''))

df4.Runs = df4.Runs.apply(lambda x : x.replace('*', ''))
```
